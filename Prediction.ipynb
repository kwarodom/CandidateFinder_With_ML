```python
import pandas as pd
```

```python
final_data = pd.read_csv("full_candidate_final.csv")
del final_data["Unnamed: 0"]
final_data
```

```python
labelData = pd.read_csv("label.csv")
myLabel = []
for i in range(len(labelData)):
    myLabel.append(labelData.iloc[i][1])
```

```python
label = myLabel.copy() # Full dato
len(final_data.columns)
len(label)
```

```python
X = final_data.copy()
X.insert(0, "label", 0, allow_duplicates=False)
for i in range(len(X)):
    X.iloc[i] = X.iloc[i].set_value("label", label[i])
X
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn import metrics
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.neural_network import MLPClassifier
from sklearn import svm
from sklearn import preprocessing
%matplotlib inline

data_x = X.copy()
y = data_x["label"]

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.30,
                                                    random_state=None)

# logmodel = LogisticRegression()
logmodel = svm.SVC()
# logmodel = MLPClassifier(solver='lbfgs',activation='relu', alpha=0.001,hidden_layer_sizes=(400,30,4,1), random_state=None, max_iter=200)

print("## Train ##")

normalized_X = preprocessing.normalize(X_train)
standardized_X = preprocessing.scale(normalized_X)


logmodel.fit(standardized_X,y_train)

predictions = logmodel.predict(standardized_X)
print(classification_report(y_train,predictions))
print(metrics.accuracy_score(y_train, predictions))    

print("")
print("## Test ##")

normalized_X_test = preprocessing.normalize(X_test)
standardized_X_test = preprocessing.scale(normalized_X_test)

predictions_new = logmodel.predict(standardized_X_test)
print(classification_report(y_test,predictions_new))
print(metrics.accuracy_score(y_test, predictions_new))    
```
